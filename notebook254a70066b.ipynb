## Dataset Overview

The dataset contains information about movies, including details such as the poster URL, title, year of release, certificate, duration, genre, rating, metascore, director, cast, votes, description, review count, review title, and review text. Below is a brief description of each column:

- **Poster**: URL to the movie poster.
- **Title**: Title of the movie.
- **Year**: Release year of the movie.
- **Certificate**: Age certificate or rating.
- **Duration (min)**: Duration of the movie in minutes.
- **Genre**: Genre(s) of the movie.
- **Rating**: IMDb rating of the movie.
- **Metascore**: Metascore rating of the movie.
- **Director**: Director of the movie.
- **Cast**: Main cast members of the movie.
- **Votes**: Number of votes the movie has received.
- **Description**: Brief description of the movie.
- **Review Count**: Number of reviews.
- **Review Title**: Title of the review.
- **Review**: Text of the review.

```python
data['Duration']  =  data['Duration (min)'].fillna(data['Duration (min)'] .mean())

data['Metascore']  =  data['Metascore'].fillna(data['Metascore'] .mean())

data['Year'] = data['Year'].fillna(0) 
data = data[data['Year'] > 0]
data['Year'] = pd.to_datetime(data['Year'].astype(str), errors='coerce')
data['Year'] = data['Year'].dt.year

data['ReviewCount'] = data['Review Count'].replace(',','', regex=True)
data['ReviewCount'] = data['ReviewCount'].fillna(0).astype(int)

data['Votes'] = data['Votes'].replace(',','', regex=True)
data['Votes'] = data['Votes'].fillna(data['Votes'].astype(float).mean()).astype(float)

pattern = r'-(.*?)\.com'
data['platform'] = data['Poster'].str.extract(pattern, expand=False)
data = data.dropna(subset=['Review'])
```

```python
data['Certificate'].unique()
```

```python
def clean_certificate(cert):
    if pd.isna(cert):
        return 'Unknown'
    cert = cert.strip().upper()
    if cert in ['R', 'PG-13', 'PG', 'U', 'U/A', 'G', 'NC-17', 'M/PG', '18+', '12+', 'PG-13', '15+']:
        return cert
    if 'UA' in cert or 'U/A' in cert:
        return 'U/A'
    if 'ALL' in cert:
        return 'All'
    if 'BANNED' in cert:
        return 'Banned'
    return 'Other' 
data['Certificate'] = data['Certificate'].apply(clean_certificate)
```

```python
data['Genre'][:10]
```

```python
splitted_genres = data['Genre'].str.split(', ', expand = True).iloc[:,:2]
```

```python
splitted_genres.columns = [f"genre{i+1}" for i in range(splitted_genres.shape[1])]
data = pd.concat([data, splitted_genres],axis=1)
```

```python
data[['lead_actor', 'second_actor', 'third_actor']] = data['Cast'].str.split(',', expand=True).iloc[:, :3]
data['reviewText'] = data[['Review Title', 'Review']].astype(str).agg(' '.join, axis=1)
```

```python
data = data.drop(columns =['Poster','Duration (min)','Cast','Review Title','Review','Genre'])
```

```python
grouped_date = data.groupby('Year').size().reset_index(name='Count')
grouped_date.max(), grouped_date.min()
```

```python
fig = go.Figure(data=[go.Scatter(x=grouped_date['Year'], y=grouped_date['Count'], mode='lines+markers')])

# Update layout for better visuals
fig.update_layout(
    title='Count of Records per Year',
    xaxis_title='Year',
    yaxis_title='Count',
    template='plotly_dark',  
    xaxis=dict(
    
        dtick=1,            
        tickangle=-45       
    )
)

# Show the plot
fig.show()
```

```python
dir_grouped =  data.groupby('Director').agg({
    'Rating': 'mean',
    'Metascore': 'mean',
    'Votes': 'sum'
}).reset_index().sort_values(by=['Rating', 'Votes'],ascending=[False, False])


top_50_directors = dir_grouped.head(50).sort_values(by='Rating', ascending=True)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(
        y=top_50_directors['Director'],
        x=top_50_directors['Rating'],
        orientation='h',  # Horizontal bars
        marker=dict(color=top_50_directors['Rating']),
        name='Average Rating'
    )
)

# Update layout for better visuals
fig.update_layout(
    title='Top 50 Directors by Average Rating',
    xaxis_title='Average Rating',
    yaxis_title='Director',
    coloraxis_colorbar=dict(title='Average Rating'),  
    
    height=1250
)

# Show the plot
fig.show()
```

```python
top_50_directors = dir_grouped.head(50).sort_values(by='Votes', ascending=True)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(
        y=top_50_directors['Director'],
        x=top_50_directors['Votes'],
        orientation='h',  # Horizontal bars
        marker=dict(color=top_50_directors['Votes']),
        name='Average Votes'
    )
)

# Update layout for better visuals
fig.update_layout(
    title='Top 50 Directors by Average Votes',
    xaxis_title='Average Votes',
    yaxis_title='Director',
    coloraxis_colorbar=dict(title='Average Votes'),  
    
    height=1250
)

# Show the plot
fig.show()
```
